```python
df_language_have_worked_with = pd.DataFrame(columns=['LanguageHaveWorkedWith', 'total'])

for index, row in df.iterrows():
        index_language = 0
        language_lst = df['LanguageHaveWorkedWith'][index].split(";")
        for i in language_lst:
            if df_language_have_worked_with['LanguageHaveWorkedWith'].isin([i]).any():
                df_language_have_worked_with.iloc[index, 1] = df_language_have_worked_with.iloc[index, 1] + 1
                print(df_language_have_worked_with.iloc[index, 1])
            else:
                #print(i)
                new_row = pd.Series([i, 0], index=['LanguageHaveWorkedWith', 'total']) 
                df_language_have_worked_with = pd.concat([df_language_have_worked_with, pd.DataFrame([new_row])], ignore_index=True)
df_language_have_worked_with.head(20)
```
